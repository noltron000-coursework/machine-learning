```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
import numpy

def keyword_sklearn(docs, k):
	vectorizer = TfidfVectorizer(stop_words='english')
	matrix = vectorizer.fit_transform(docs)
	scores = numpy.sum(matrix, axis=0)
	print(scores)
	scores = numpy.ravel(scores)
	print(scores)
	# get all words
	words = vectorizer.get_feature_names()
	print(words)
	my_dict = dict(zip(words, scores))

	sorted_dict = sorted(
		my_dict.items(),
		key=lambda word_info: word_info[1],
		reverse=True
	)

	return sorted_dict

documents = [
	'The sky is blue',
	'The sun is bright',
	'The sun in the sky is bright',
	'we can see the shining sun, the bright sun'
]

print(keyword_sklearn(documents, 3))
```

```python
import math
import codecs

def multidimensional_pythagorean(A, B):
	assert(len(A) == len(B))
	y = 0
	for a_val, b_val in zip(A, B):
		x = a_val - b_val
		y = math.sqrt(x**2 + y**2)
	return y

def get_distances_of_words(text_file, *words):
	words = set(words)
	data_dict = {}
	for idx, row in enumerate(text_file):
		data = row.split()
		word = data[0]
		data = data[1:]
		if word in words:
			words.remove(word)
			data_dict[word] = data
		if len(words) == 0:
			break
	return data_dict

def get_distances(*words):
	with codecs.open('./ignore/glove.840B.300d.txt', 'r') as text_file:
		data_dict = get_distances_of_words(text_file, *words)
	print(data_dict)

print(get_distances('pizza', 'running'))

A = [1,1,1,1]
B = [2,2,2,2]
# print(multidimensional_pythagorean(A,B))
```
